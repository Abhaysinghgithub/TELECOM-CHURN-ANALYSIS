```python
s1=df['State'].unique()
s2=df.groupby(['State'])['Churn'].mean()
plt.figure(figsize=(15,8))
plt.plot(s1,s2,color='r',marker='o')
plt.title('State churn rate')
plt.xlabel('State',fontsize=15)
plt.ylabel('churn rate',fontsize=15)
```

```python
plt.figure(figsize=(15,8))
(df.groupby(['State'])['Churn'].mean()*100).sort_values(ascending=False).head(10).plot.bar(color='r')
plt.xlabel('state' ,fontsize=15)
plt.ylabel('percntage',fontsize=15)
```

**Calculate state vs Churn percntage**

```python
State_data= pd.crosstab(df["State"],df["Churn"])
State_data['percentage_churn']= State_data.apply(lambda x : x[1]*100/(x[0]+x[1]),axis=1)
State_data
```

```python
(df.groupby(['State'])['Churn'].mean()).sort_values(ascending=False).head(10)
```

```python
Area_code_data= pd.crosstab(df["Area code"],df["Churn"])
Area_code_data['percntage_churn']= Area_code_data.apply(lambda x : x[1]*100/(x[0]+x[1]),axis=1)
Area_code_data
```
